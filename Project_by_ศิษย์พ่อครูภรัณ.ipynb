#การเชื่อมต่อข้อมูล

```python
import os
import pandas as pd

# กำหนด paths ที่ต้องการใช้
paths = [path1, path2, path3, path4, path5, path6, path7, path8]
dataframes = []  # ใช้สำหรับเก็บ DataFrame ที่อ่านมา

# อ่านข้อมูลจากแต่ละ directory
for path in paths:
    try:
        # ตรวจสอบและอ่านไฟล์ CSV ทุกไฟล์ในแต่ละ path
        for file_name in os.listdir(path):
            if file_name.endswith('.csv'):
                file_path = os.path.join(path, file_name)
                df = pd.read_csv(file_path)

                # เพิ่มคอลัมน์ 'City' โดยการดึงชื่อเมืองจาก file_name
                city_name = file_name.replace('.csv', '').split('-stop-and-search')[0].split('-', 2)[-1]
                df['City'] = city_name

                # เพิ่ม DataFrame ลงใน list
                dataframes.append(df)
    except FileNotFoundError:
        print(f"Warning: Directory not found at {path}")
    except Exception as e:
        print(f"An error occurred while reading files from {path}: {e}")

# ตรวจสอบว่า dataframes มีข้อมูลไหม
if dataframes:
    # รวมข้อมูลจากทุก DataFrame
    all_data = pd.concat(dataframes, ignore_index=True)
    print("Data read from folders and combined successfully.")
else:
    print("No data loaded from the specified paths.")

# ผลลัพธ์ทั้งหมดจะอยู่ใน all_data
```

#การทำความสะอาดข้อมูล (Data Cleaning)

```python
import pandas as pd

# ตัวอย่าง DataFrame (สมมติว่า df คือ DataFrame ของคุณ)
# df = pd.read_csv('your_data.csv')

# Replace missing values for categorical columns with "Unknown" or the mode
categorical_columns = ['Gender', 'Self-defined ethnicity', 'Officer-defined ethnicity', 'Outcome', 'Object of search', 'Part of a policing operation', 'Policing operation', 'Outcome linked to object of search', 'Removal of more than just outer clothing']
for col in categorical_columns:
    # Use mode for imputation if possible, otherwise use 'Unknown'
    try:
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value)
    except:
        df[col] = df[col].fillna('Unknown')

# Remove rows where Latitude or Longitude are missing
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# Replace missing values in the Date column with the latest date
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # แปลงให้เป็น datetime
df['Date'] = df['Date'].fillna(df['Date'].max())


# Check the result
print(df.isnull().sum())  # Check the number of missing values after imputation
```

```python
# แทนค่าค่าว่างใน 'Age range' และ 'Legislation'
df['Legislation'] = df['Legislation'].fillna('Unknown')  # ใช้ "Unknown" หากไม่มีค่า mode ที่เหมาะสม
df['Age range'] = df['Age range'].fillna('Unknown')
# ตรวจสอบผลลัพธ์
print(df.isnull().sum())  # ตรวจสอบจำนวน missing values หลังจากการแทนค่า
```

```python
# Extract the month from the 'Date' column
df['Month'] = df['Date'].dt.to_period('M')

# Get the complete list of unique cities
all_cities = set(df['City'].unique())

# Group by month and get the unique cities in each month
cities_by_month = df.groupby('Month')['City'].unique()

print("Number of unique cities per month:")
for month, cities in cities_by_month.items():
    print(f"{month}: {len(cities)} cities")

print("\nMissing cities per month:")
for month, cities in cities_by_month.items():
    missing_cities = all_cities - set(cities)
    if missing_cities:
        print(f"{month}: {missing_cities}")
    else:
        print(f"{month}: None missing")
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

# Treemap of Object Categories
fig_treemap = px.treemap(df, path=[px.Constant("all"), 'Object_category'],
                  title='Distribution of Object Categories')
fig_treemap.show()

# Stacked Bar Chart of Object Categories by Region
# First, count the occurrences of each object category within each region
object_category_by_region = df.groupby(['Region', 'Object_category']).size().reset_index(name='Count')

# Create the stacked bar chart
fig_stacked_bar = px.bar(object_category_by_region, x='Region', y='Count', color='Object_category',
                         title='Object Categories by Region',
                         labels={'Count': 'Number of Searches', 'Region': 'Region', 'Object_category': 'Object Category'})
fig_stacked_bar.show()
```

```python
# Create a histogram of the 'Age range' column
fig_hist = px.histogram(df, x='Age range', title='Distribution of Age Range in Stop and Search')
fig_hist.show()
```

```python
import plotly.express as px
import pandas as pd

# --- Radar Chart ---
# Aggregate data for Radar Chart
# We need counts of each Object_category per Region
radar_data = df.groupby(['Region', 'Object_category']).size().reset_index(name='Count')

# To make it suitable for a radar chart, we need to pivot the table
# regions will be the lines, object categories will be the axes
radar_pivot = radar_data.pivot(index='Region', columns='Object_category', values='Count').fillna(0).reset_index()

# Normalize the data for better comparison on the radar chart (optional but recommended)
# This scales the counts for each object category between 0 and 1 across all regions
for col in radar_pivot.columns:
    if col != 'Region':
        radar_pivot[col] = radar_pivot[col] / radar_pivot[col].max()

# Melt the pivoted data back to long format for Plotly Express line_polar
radar_melted = radar_pivot.melt(id_vars='Region', var_name='Object_category', value_name='Normalized_Count')


# Create the radar chart
fig_radar = px.line_polar(radar_melted, r='Normalized_Count', theta='Object_category',
                          line_close=True,
                          color='Region',
                          title='Object Category Distribution by Region (Normalized)')
fig_radar.update_traces(fill='toself')
fig_radar.show()


```
